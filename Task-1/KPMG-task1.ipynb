#### Data Cleaning on Transactions Data

```python
trans.isna().sum()
```

```python
trans["online_order"]=trans["online_order"].fillna("False")
```

```python
trans["brand"]=trans["brand"].fillna("others")
```

```python
trans.product_line.value_counts()
```

```python
trans["product_line"]=trans["product_line"].fillna(trans.product_line.mode()[0])
```

```python
trans.product_line.value_counts()
```

```python
trans.product_class.value_counts()
```

```python
trans.product_class=trans.product_class.fillna(trans.product_class.mode()[0])
```

```python
trans.product_size=trans.product_size.fillna("others")
```

```python
trans.standard_cost=trans.standard_cost.fillna(trans.standard_cost.mean())
```

```python
trans=trans.drop("product_first_sold_date",axis=1)
```
